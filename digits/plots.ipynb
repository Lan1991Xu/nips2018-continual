```python
import pandas as pd
import glob

import seaborn as sns
import matplotlib.pyplot as plt
import re

import h5py

from analysis import load_results, plot_cfm


%matplotlib inline
```

# Generate Plots for Domain Adaptation Experiments

```python
df = pd.DataFrame(load_results("results/noiseadapt_predictions.hdf5"), columns = ["source", "algorithm", "acc"])
df = df[df.algorithm != "DANNSolver"]
results = df.pivot("algorithm", "source", "acc")
```

```python
sns.set_context('poster')
plt.figure(figsize=(7,5))
    
plot_cfm(results, vmin=65, vmax=100)
plt.savefig('results/noise-adapt-comparison.pdf', bbox_inches='tight')
plt.show()
```
